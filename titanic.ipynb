```python
import pandas as pd
df = pd.read_csv("train.csv")  # For CSV files
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=df, palette="coolwarm")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution of Titanic Passengers")
plt.show()
```

Interpretation: You’ll see how many males and females were on board.

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

Interpretation: This will show whether men or women had a higher survival chance.



3.



**Patterns in Data**

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Survival by Age")
plt.show()
```

This helps see which age groups had higher survival rates.
-  Children had a higher chance of survival.
-  Older passengers were more likely to not survive.
- Age alone wasn’t enough to predict survival.



```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

This helps see which age groups had higher survival rates.

**Conclusion:**

From the analysis, we find that:

*   Women survived more than men.
*   Younger passengers had a higher survival rate.
*   First-class passengers had a better chance of survival.
